```python
#Importing Required Libraries and Modules

from nltk.util import pr
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
import nltk 
nltk.download('stopwords')
```

```python
#Loading Dataset

data = pd.read_csv("/content/drive/MyDrive/6th_SEM_MINI_PROJECT/twitter.csv")
print(data.head())
```

```python
#Defining Labels

data["labels"] = data["class"].map({0: "Hate Speech", 1: "Offensive Language", 2: "No Hate and Offensive"})
print(data.head())
```

```python
#Labelling Tweets with Labels

data = data[["tweet", "labels"]]
print(data.head())
```

```python
#Importing RE Module and NLTK

import re
import nltk
```

```python
#Stemming

stemmer = nltk.SnowballStemmer("english")
```

```python
#Collecting Stop-words for English Language
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))
print (stopword)
```

```python
#Cleaning Data and Focusing on main words

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
data["tweet"] = data["tweet"].apply(clean)
print(data.head())
```

```python
#Generating Seperate Arrays for tweet and labels

x = np.array(data["tweet"])
y = np.array(data["labels"])
print(x)
print("\n")
print(y)
```

```python
#Extracting features from text

cv = CountVectorizer()
X = cv.fit_transform(x)
print(X)
```

```python
#Splitting Dataset using train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train)
print("\n")
print(X_test)
print("\n")
print(y_train)
print("\n")
print(y_test)
print("\n")
```

```python
#Measuring quality of Splitting Dataset

clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
#Detecting Hate Speech

def hate_speech_detection():
  
    user = input("Enter any Tweet: ")
    if len(user) < 1:
        print(" ")
    else:
        sample = user
        data = cv.transform([sample]).toarray()
        a = clf.predict(data)
        print(a)
hate_speech_detection()
```
